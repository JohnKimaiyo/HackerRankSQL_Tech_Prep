```python
import pandas as pd
import random
from faker import Faker
from pathlib import Path

# Initialize Faker
fake = Faker()

# Output folder (creates folder if not exists)
output_dir = Path("sql_interview_datasets")
output_dir.mkdir(exist_ok=True)

# -------------------- STUDENTS --------------------
students = pd.DataFrame({
    "student_id": range(1, 101),
    "name": [fake.name() for _ in range(100)],
    "marks": [random.randint(35, 100) for _ in range(100)]
})
students.to_csv(output_dir / "students.csv", index=False)

# -------------------- DEPARTMENTS --------------------
departments = pd.DataFrame({
    "department_id": range(1, 6),
    "department_name": [
        "Finance", "Technology", "HR", "Sales", "Operations"
    ]
})
departments.to_csv(output_dir / "departments.csv", index=False)

# -------------------- EMPLOYEES --------------------
employees = pd.DataFrame({
    "employee_id": range(1, 101),
    "employee_name": [fake.name() for _ in range(100)],
    "salary": [random.randint(30000, 150000) for _ in range(100)],
    "department_id": [random.randint(1, 5) for _ in range(100)]
})
employees.to_csv(output_dir / "employees.csv", index=False)

# -------------------- CUSTOMERS --------------------
customers = pd.DataFrame({
    "customer_id": range(1, 101),
    "customer_name": [fake.name() for _ in range(100)],
    "country": [fake.country() for _ in range(100)]
})
customers.to_csv(output_dir / "customers.csv", index=False)

# -------------------- PRODUCTS --------------------
products = pd.DataFrame({
    "product_id": range(1, 21),
    "product_name": [fake.word().capitalize() for _ in range(20)],
    "price": [round(random.uniform(10, 500), 2) for _ in range(20)]
})
products.to_csv(output_dir / "products.csv", index=False)

# -------------------- ORDERS --------------------
orders = pd.DataFrame({
    "order_id": range(1, 201),
    "customer_id": [random.randint(1, 100) for _ in range(200)],
    "order_date": [
        fake.date_between(start_date="-2y", end_date="today")
        for _ in range(200)
    ]
})
orders.to_csv(output_dir / "orders.csv", index=False)

# -------------------- ORDER ITEMS --------------------
order_items = pd.DataFrame({
    "order_item_id": range(1, 401),
    "order_id": [random.randint(1, 200) for _ in range(400)],
    "product_id": [random.randint(1, 20) for _ in range(400)],
    "quantity": [random.randint(1, 10) for _ in range(400)]
})
order_items.to_csv(output_dir / "order_items.csv", index=False)

print("‚úÖ CSV datasets generated successfully!")
print(f"üìÅ Location: {output_dir.resolve()}")
```
